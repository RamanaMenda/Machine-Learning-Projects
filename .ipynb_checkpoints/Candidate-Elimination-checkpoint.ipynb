```python
import pandas as pd
import numpy as np
from copy import deepcopy
```

```python
data = pd.read_csv('EnjoySport.csv')
```

```python
print(data)
```

```python
def candidateElimination(s,g):
    n = len(data.keys())
    
    for d in data.iterrows():
        if d[1][-1] == 'Yes':
            for i in range(n-1):
                if s[i] != '?':
                    if s[i] is None:
                        s[i] = d[1][i]
                    elif s[i] != d[1][i]:
                        s[i] = '?'
            newg = []
            for i in g:
                for j in range(n-1):
                    if i[j] != s[j] and i[j] != '?':
                        break
                else:
                    newg.append(i)
            g = newg
        else:
            newg = []
            for i in g:
                for j in range(n-1):
                    if d[1][j] != s[j] and s[j] != '?':
                        newi = deepcopy(i)
                        newi[j] = s[j]
                        newg.append(newi)
            g = newg
        print(s,g)
    return s, g       
```

```python
n = len(data.keys())
s = [None for i in range(n-1)]
g = [['?' for i in range(n-1)]]
candidateElimination(s,g)
```
